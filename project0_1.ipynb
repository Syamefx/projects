# **Algorithmüëá**

Step 1: `Start`

Step 2: **`Initialize Variables`**

Create an empty list expenses to store data.

Set budget limit (e.g., 1000).

Set total_spent to 0.

Create an empty dictionary category_totals to track spending per category.

Set highest_amount to 0.

Step 3: `Input Data (Data Entry Loop)`

Loop continuously:

Read date, category, and amount from the user.

Store these three values as a list [date, category, amount].

Append this list to the main expenses list.

Ask user if they want to add another. If "no", break the loop.

Step 4: `Analyze Data (Processing Loop)`

Iterate through each entry in the expenses list:

Add entry[amount] to total_spent.

Check Category:

If category exists in category_totals, add amount to it.

Else, create the key in category_totals and set it to the amount.

Check Highest Spend:

If entry[amount] > highest_amount:

Update highest_amount = entry[amount].

Store the date and category of this expense.

Step 5:` Find Dominant Category`

Initialize max_cat_amount to 0 and dominant_category to empty.

Iterate through category_totals:

If current category total > max_cat_amount, update dominant_category and max_cat_amount.

Step 6:` Generate Output`

Print total_spent.

Budget Check:

If total_spent > budget, print "Budget Breached" and the difference.

Else, print "Within Budget".

Print breakdown of spending by category.

Print the highest_amount (with date/category) and the dominant_category.

Step 7: `Stop`







```python
# [Date, Category, Amount]
expenses = [
    ['2023-10-01', 'Food', 150.00],
    ['2023-10-02', 'Transport', 50.00],
    ['2023-10-05', 'Food', 200.00],
    ['2023-10-10', 'Bills', 500.00],
    ['2023-10-15', 'Entertainment', 300.00],
    ['2023-10-20', 'Food', 100.00],
    ['2023-10-25', 'Bills', 120.00]
]

budget = 1000.00  # Our monthly limit
```

```python
expenses = []
while True:
    print("\n--- Add New Expense ---")
    date = input("Enter Date (YYYY-MM-DD): ")
    category = input("Enter Category (Food, Travel, Bills, etc.): ")

    amount = float(input("Enter Amount: "))

    entry = [date, category, amount]

    expenses.append(entry)


    more = input("Add another? (yes/no): ")
    if more.lower() != 'yes':
        break

print("\nData Entry Complete!")
print("Total expenses logged:", len(expenses))
```

```python
# --- Initializing Variables ---
total_spent = 0
category_totals = {}  # A dictionary to hold "Category": Total Amount
highest_amount = 0
highest_expense_date = ''
highest_expense_category = ''

# --- The Main Loop ---
for entry in expenses:
    # Unpack the list for easier reading
    date = entry[0]
    category = entry[1]
    amount = entry[2]

    # 1. Update Total Spending
    total_spent = total_spent + amount

    # 2. Update Category Totals
    if category in category_totals:
        category_totals[category] += amount
    else:
        category_totals[category] = amount

    # 3. Check for Highest Expense (Pattern Detection)
    if amount > highest_amount:
        highest_amount = amount
        highest_expense_date = date
        highest_expense_category = category

# --- Finding the Dominant Category ---
# We loop through the dictionary we just built to find the max
dominant_category = ''
max_cat_amount = 0

for cat in category_totals:
    if category_totals[cat] > max_cat_amount:
        max_cat_amount = category_totals[cat]
        dominant_category = cat
```

```python
print("\n--- Expense Report ---")
print("Total Spent: $", total_spent)

# Budget Alert Logic
if total_spent > budget:
    print(f"‚ö†Ô∏è ALERT: You have exceeded your budget of ${budget}!")
    print(f"Overspend amount: ${total_spent - budget}")
else:
    print(f"‚úÖ You are within your budget. Remaining: ${budget - total_spent}")

print("\n--- Category Breakdown ---")
for cat in category_totals:
    print(f"{cat}: ${category_totals[cat]}")

print("\n--- Insights ---")
print(f"Highest Single Spend: ${highest_amount} on {highest_expense_date} ({highest_expense_category})")
print(f"Dominant Spending Category: {dominant_category} (Total: ${max_cat_amount})")
```
