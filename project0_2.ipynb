# ***`Question 7`***

> Quiz Engine with Dynamic Scoring




# ***Algorithmüëá***

Step 1: `Start`

Step 2: `Initialize Data Structures`

Create question_bank: A list of dictionaries, where each dictionary contains Question, Options, Correct Answer, and Difficulty.

Step 3: `Pseudo-Random Shuffle (No Library)`

Input a seed number from the user.

Create a list of indices [0, 1, 2, ... n].

Iterate through the indices:

Calculate a swap position using the math formula: (seed * constant + increment) % total_questions.

Swap the current index with the calculated position index.

Step 4: `Start Quiz Execution`

Initialize score = 0, correct_count = 0, wrong_count = 0.

Iterate through the shuffled indices list:

Retrieve the question data corresponding to the current index.

Display: Print the Question, Difficulty Level, and Options.

Input: Read user answer (apply string handling: trim spaces, convert to lowercase).

Step 5: `Evaluate Answer`

Compare user input with the correct answer.

If Correct:

Check Difficulty:

If "Easy", add 10 to score.

If "Medium", add 20 to score.

If "Hard", add 30 to score.

If Wrong:

Deduct 5 from score (Negative Marking).

Display the correct answer.

Step 6: `Generate Report`

Print score.

Calculate and print Accuracy % (correct_count / total_questions * 100).

Verdict:

If score > threshold (e.g., 50), print "Pass/Excellent".

Else, print "Fail/Practice More".

Step 7: `Stop`




```python
# Format:
# {
#   "q": "Question Text",
#   "opt": ["Option A", "Option B", "Option C", "Option D"],
#   "ans": "Correct Option String",
#   "diff": "Easy/Medium/Hard"
# }

question_bank = [
    {
        "q": "What is the capital of France?",
        "opt": ["Berlin", "Madrid", "Paris", "Rome"],
        "ans": "Paris",
        "diff": "Easy"
    },
    {
        "q": "Which data structure uses LIFO?",
        "opt": ["Queue", "Stack", "List", "Tree"],
        "ans": "Stack",
        "diff": "Medium"
    },
    {
        "q": "What is 2 ** 3 in Python?",
        "opt": ["6", "8", "9", "5"],
        "ans": "8",
        "diff": "Easy"
    },
    {
        "q": "Which keyword is used for functions?",
        "opt": ["func", "def", "define", "function"],
        "ans": "def",
        "diff": "Easy"
    },
    {
        "q": "What is the complexity of binary search?",
        "opt": ["O(n)", "O(log n)", "O(n^2)", "O(1)"],
        "ans": "O(log n)",
        "diff": "Hard"
    }
]
```

```python
def get_shuffled_indices(total_items, seed):
    indices = []
    # Create a list of standard indices [0, 1, 2, 3, 4]
    for i in range(total_items):
        indices.append(i)

    # A simple manual shuffle using the seed
    # We swap items based on the seed value
    for i in range(total_items):
        seed = (seed * 11 + 7) % total_items # Simple math to get a new position

        # Swap the current item with the calculated position
        temp = indices[i]
        indices[i] = indices[seed]
        indices[seed] = temp

    return indices
```

```python
print("--- Welcome to the Dynamic Quiz Engine ---")
seed_input = int(input("Enter a lucky number to start: "))

# 1. Get the random order
question_order = get_shuffled_indices(len(question_bank), seed_input)

# 2. Initialize Score
total_score = 0
correct_count = 0
wrong_count = 0

# 3. Start the Loop
print("\nStarting Quiz...\n")

for index in question_order:
    # Extract the question data using the shuffled index
    current_q = question_bank[index]

    print(f"Difficulty: {current_q['diff']}")
    print(f"Q: {current_q['q']}")

    # Display Options (String Handling)
    options = current_q['opt']
    print(f"A) {options[0]}   B) {options[1]}   C) {options[2]}   D) {options[3]}")

    user_ans = input("Your Answer (type the word): ").strip() # Remove extra spaces

    # Check Answer
    # We use .lower() for case-insensitive string comparison
    if user_ans.lower() == current_q['ans'].lower():
        print("‚úÖ Correct!")
        correct_count += 1

        # Dynamic Scoring Logic
        if current_q['diff'] == 'Easy':
            total_score += 10
        elif current_q['diff'] == 'Medium':
            total_score += 20
        else: # Hard
            total_score += 30

    else:
        print(f"‚ùå Wrong! The correct answer was: {current_q['ans']}")
        wrong_count += 1
        total_score -= 5 # Negative marking

    print("-" * 30) # Separator line

# 4. Final Report
print("\n--- Final Report Card ---")
print(f"Total Score: {total_score}")
print(f"Correct Answers: {correct_count}")
print(f"Wrong Answers: {wrong_count}")

# Avoid division by zero error if list is empty
if len(question_bank) > 0:
    accuracy = (correct_count / len(question_bank)) * 100
    print(f"Accuracy: {accuracy}%")

if total_score > 50:
    print("Verdict: Excellent Job!")
else:
    print("Verdict: Keep Practicing.")
```
